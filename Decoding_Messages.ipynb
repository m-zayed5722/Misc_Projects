```python
from google.colab import files

def read_lines_from_uploaded_file():
    uploaded = files.upload()

    if not uploaded:
        print("No file uploaded.")
        return None

    file_name = list(uploaded.keys())[0]

    try:
        with open(file_name, 'r') as file:
            lines = file.readlines()
            words_list = [line.strip().split() for line in lines]
            return words_list
    except FileNotFoundError:
        print(f"File not found: {file_name}")
        return None

# Example usage:
result = read_lines_from_uploaded_file()

if result is not None:
    print(result)
```

```python
from google.colab import files

def create_pyramid_and_get_last_numbers():
    uploaded = files.upload()

    if not uploaded:
        print("No file uploaded.")
        return None

    file_name = list(uploaded.keys())[0]

    try:
        with open(file_name, 'r') as file:
            lines = file.readlines()
            pyramid = []
            last_numbers = []

            for line in lines:
                items = line.strip().split()
                numbers = [int(item) if item.isdigit() else item for item in items]
                pyramid.append(numbers)

            for i in range(len(pyramid)):
                last_number = pyramid[i][-1]
                last_numbers.append(last_number)
                print(" ".join(map(str, pyramid[i])))

            return last_numbers
    except FileNotFoundError:
        print(f"File not found: {file_name}")
        return None

# Example usage:
result = create_pyramid_and_get_last_numbers()

if result is not None:
    print("Last numbers:", " ".join(map(str, result)))
```
